Supongamos que para un evento dado, el tamao de la distribucin de prdidas (gravedad) de nuestra lnea de negocio se describe mejor mediante una distribucin gamma de $\alpha$ = 7 y $\lambda$ = 0.0025.\\
La distribucin gamma ($\Gamma$ como un escalar) con parmetros forma = $\lambda$ y escala = $\alpha$ tiene como funcin de densidad:
\begin{equation*}
f(x) = \lambda e^{-\lambda x} \frac{(\lambda x)^{\alpha-1}}{\Gamma(\alpha)}
\end{equation*}
El valor esperado (media) y la varianza de una variable aleatoria x de distribucin gamma son:
\begin{equation*}
\textbf{E(x)} = \alpha \theta = \alpha / \lambda \ \ \ y \ \ \ \textbf{Var(x)} = \alpha \theta^2 = \alpha / \lambda^2 %\sigma^2
\end{equation*}
Se pide:
* Calcular la media y la desviacin estndar de la siniestralidad total

```python
import numpy as np
import scipy.stats as st

alpha = 7.
lamdba = 0.0025
theta = 1/lamdba
```

```python
# Opcion 1:
mean = alpha / lamdba
var = alpha / (lamdba ** 2)
std = np.sqrt(var)
print(mean, std)
```

*En Scipy (st.gamma), los parmetros de la gamma son: a=alpha y scale=1/lamdba (theta)*

```python
# Opcion 2:
print(st.gamma.mean(a=alpha, scale=1/lamdba), st.gamma.std(a=alpha, scale=1/lamdba))
```

```python
# Opcion 3:
siniestralidad = st.gamma(a=alpha, scale=1/lamdba)
print(siniestralidad.mean(), siniestralidad.std())
```

```python
# Opcion 4: 
m, v = siniestralidad.stats('mv')
print(m,np.sqrt(v))
```

* Calcular la siniestralidad esperada en los siguientes percentiles: 90th, 95th, 99th, 99.6th y 99.8th 

*Los percentiles de una distribucin corresponde con el .ppf (funcin de punto porcentual). 
Ejemplo: st.gamma.ppf(0.5,a=alpha,scale=1/lambda)*

```python
percentiles = (0.5, 0.9, 0.95, 0.99, 0.996, 0.998)

for i in percentiles:
    print('El percentil %.2f teorico es %.2f' % (i, st.gamma.ppf(i, a=alpha, scale=1/lamdba)))
```

* Aplicar 10.000 simulaciones aleatorias con numpy (np.random.gamma) y calcular los anteriores percentiles.

*Los parmetros de numpy son np.random.gamma(alpha, 1/lamdba, size=10000)*

```python
s = np.random.gamma(alpha, theta, size=10000)
for i in percentiles:
    # los percentiles en np.percentile van de 0 a 100
    print('El percentil %.2f de la simulacion es %.2f' % (i,(np.percentile(s, i*100))))
```

Mostramos la grfica de la distribucin:

```python
import matplotlib.pyplot as plt

plt.hist(s, 50, density=True)
x = np.linspace (0, 9000, 200) 
y = st.gamma.pdf(x, a=alpha, scale=1/lamdba)
plt.plot(x, y)
plt.show()
```
