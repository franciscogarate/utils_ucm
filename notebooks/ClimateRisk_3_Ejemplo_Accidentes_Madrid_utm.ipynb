```python
import pandas as pd
import numpy as np
from folium import Map, plugins
from folium.plugins import HeatMap
import utm
```

https://visualizadatos.madrid.es/pages/accidentes-de-trafico

```python
map_madrid = Map(location=[40.41686, -3.7041056], zoom_start = 13)
map_madrid
```

```python
df = pd.read_excel('./utils_ucm/data/AccidentesBicicletas_2024.xlsx', usecols=(15,16))

def utm2latlon(x,y):
    try:
        return utm.to_latlon(x, y, 30, 'U')
    except:
        return (0,0)

df['coordenada_x_utm'] = df['coordenada_x_utm'].astype(float)
df['coordenada_y_utm'] = df['coordenada_y_utm'].astype(float)
df['coord'] = df.apply(lambda x: utm2latlon(x.coordenada_x_utm, x.coordenada_y_utm), axis=1)
```

```python
coord_acc = np.asarray(df.coord)
```

```python
HeatMap(coord_acc, radius=20).add_to(map_madrid)
map_madrid
```
