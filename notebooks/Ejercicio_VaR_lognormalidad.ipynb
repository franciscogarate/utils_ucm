```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt

gaussian = st.norm(loc=4.0, scale=1.0) #media = 4.0, desviacion estandar = 1.0
x = np.linspace(0.0,8.0,100)
y = gaussian.pdf(x)
plt.plot(x,y)
```

```python
a = st.norm.cdf(1.65)
b = st.norm.ppf(0.995) #percentil
print(a, b)
```

Grafico de la distribucion lognormal, media = 4.0, desviacion estandar = 1.0

```python
lognormal = st.lognorm(s=1.0, scale=4.0) 
x = np.linspace(0.0,30,100)
y = lognormal.pdf(x)
plt.plot(x,y)
```

Para una entidad de RC Auto, cal sera un VaR(99.5%) de que el importe de la mejor estimacin de sus reservas (23.600.000) para hacer frente a la siniestralidad de los 12 siguientes meses sea suficiente?

Si $VaR(99.5\%) = Z \cdot \sigma \cdot \sqrt{t}$ , entonces:
* Z = 3
* $\sigma$ = 9\%

```python
import numpy as np

Z = 3
s = 0.09
t = 1
var = Z * s * np.sqrt(t) * 23600000
var
```
