```python
import pandas as pd
import numpy as np
from utils_ucm import lx_decesos_m
```

```python
def obtener_lx(edad):
  return lx_decesos_m[edad]
```

```python
obtener_lx(20)
```

```python
edad = 50
fval = '2024-12-31'
```

```python
fechas = pd.date_range(start='2024-12-31', periods=len(lx_decesos_m)-edad, freq='YE')
```

```python
df = pd.DataFrame(fechas, columns=['fecha'])
df.head(3)
```

```python
df['edad'] = edad + df.index
df.head(3)
```

```python
df['lx'] = df['edad'].apply(obtener_lx)
df.head(3)
```

```python
df['qx'] = df['lx'].diff(-1).fillna(0) / df['lx'][0]
df
```

```python
df['capital'] = 5000
df['pagos_esperados'] = df.capital * df.qx
df['factor_dto'] = 1/ (1+ 2.18/100) ** (df.index)
df.head()
```

```python
df.pagos_esperados @ df.factor_dto
```

```python
sum(df.pagos_esperados)
```
