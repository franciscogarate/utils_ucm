```r
# prompt: genera numeros aleatorios en base a una distribucion normal, media 20 desviación estandar 5 y graficalos en R

# Genera números aleatorios
numeros_aleatorios <- rnorm(100, mean = 20, sd = 5)

# Crea el histograma
hist(numeros_aleatorios,
     main = "Histograma de Números Aleatorios",
     xlab = "Valores",
     ylab = "Frecuencia",
     col = "lightblue",
     border = "black")
```
